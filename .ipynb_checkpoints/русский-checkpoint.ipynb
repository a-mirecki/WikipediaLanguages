## 8. выводы
Język rosyjski jest wyjątkowy, bo żaden inny język w podobnym teście nie uzyskał tak ogromnej skuteczności (drugi najlepszy wynik oscylował w granicach 93%). 6% skuteczności więcej to w kategoriach sztucznej inteligencji bardzo dużo. Oznacza to, że rosyjski najbardziej odbiega od innych popularnych języków, nawet po przełamaniu bariery innego alfabetu, a tym samym jest językiem wyjątkowym.
